```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
import math

c0 = [0,0,2.17,0]                   #l=217 cm 
tempo=np.arange(0,10,0.02)        #considera-se que o fio presso ao peso está na origem do sistema cartesiano
g = 10

def EqDif (Y,tempo):
    x = Y[0]
    Vx = Y[1]
    y = Y[2]
    Vy = Y[3]
    T = g*(x/(math.sqrt(x**2+y**2))) + (Vx**2+Vy**2)/(math.sqrt(x**2+y**2))**2 * (math.sqrt(x**2+y**2))
    dxdt = Vx
    dvxdt = g - T * (x/math.sqrt(x**2+y**2)) 
    dydt = Vy
    dvydt = -T * (y/math.sqrt(x**2+y**2))
    return [dxdt,dvxdt,dydt,dvydt]
```

```python
solucao = odeint(EqDif,c0,tempo)
       
plt.plot(solucao[:,2],-solucao[:,0])
plt.xlabel ("Posição em y")
plt.ylabel("Posição em x")
plt.axis("equal")
plt.grid()
plt.show()
```

```python
plt.plot(tempo, -solucao[:,2],)
plt.grid(True)
plt.ylabel("posição y (em metros)")
plt.xlabel("tempo (em segundos)")
plt.title("posição y em função do tempo")
plt.show()
```
